```python
#Parameter Nullen
ausgaben = {}
budget = 0
rechner_active = True

print("Willkommen beim Budgetrechner 2.0!")

#Menü Schleife
while rechner_active == True:
  print("\n\nGeben Sie '1' ein, um eine neue Ausgabe zu tracken.")
  print("Geben Sie '2' ein um ihr monatliches Budget anzupassen.")
  print("Geben Sie '3' ein, um sich genauere Statistiken anzuschauen.")
  print("Geben Sie '4' ein, um eine Ausgabe zu löschen.")
  menu_wahl = input("Geben Sie '5' ein um den Budgetrechner zu beenden. ")
  if menu_wahl.isdigit():
    menu_wahl = int(menu_wahl)
  else:
    print("\nUngültige Eingabe.")
    continue

#Neue Ausgabe Tracken
  if menu_wahl == 1:
    ausgabe = input("\nWas für eine Ausgabe möchten Sie tracken? ")
    kosten = (input("\nWie hoch war die Ausgabe? "))
    if kosten.isdigit():
      kosten = int(kosten)
      ausgaben[ausgabe.lower().strip()] = kosten
      print(f"\nDie Ausgabe {ausgabe} wurde erfolgreich getracked!")
    else:
      print("Ungültige Eingabe")

#Budget festlegen
  elif menu_wahl == 2:
    budget = input("\nWie hoch ist Ihr neues monatliches Budget? ")
    if budget.isdigit():
      budget = int(budget)
      print("Ihr Budget wurde aktualisiert.")
    else:
      print("\nUngültige Eingabe.")

#Statistik
  elif menu_wahl == 3:
    print("\nWillkommen im Statistik Menü.")
    if budget == 0:
      print("\nHinweis: Sie haben kein Budget gesetzt.")
    else:
      print(f"\nIhr Budget beträgt {budget}.")
    print("\nSie haben folgende Ausgaben getrackt:")
    for ausgabe, kosten in ausgaben.items():
      print(f"{ausgabe.title()} : {kosten}")
    print("\nIhre Gesamtausgaben belaufen sich auf:")
    gesamtausgaben = sum(ausgaben.values())
    print(gesamtausgaben)
    print("Ihr Restbudget beläuft sich auf:")
    restbudget = budget-gesamtausgaben
    print(restbudget)

#Ausgaben löschen
  elif menu_wahl == 4:
    loeschmodus = True
    while loeschmodus == True:
      loeschen = input("\nWelche Ausgabe möchten sie löschen? ")
      loeschen = loeschen.lower().strip()
      if loeschen in ausgaben.keys():
        sicher = input(f"Sicher, dass sie {loeschen} löschen wollen?")
        if sicher.lower().strip() == "ja":
          del ausgaben[loeschen]
          print(f"\nDie Ausgabe {loeschen} wurde erfolgreich gelöscht.")
        elif sicher.lower().strip() == "nein":
          print(f"\n{loeschen} wurde nicht gelöscht.")
        else:
          print("\nUngültige Ausgabe")
      else:
        print("\nDiese Ausgabe existiert leider nicht.")
      weitere = input("\nMöchten sie weitere Ausgaben löschen? ")
      if weitere.lower().strip() != "ja":
        loeschmodus = False

#Budgetrechner beenden
  elif menu_wahl == 5:
    print("\nDanke und schönen Tag!")
    break
```
