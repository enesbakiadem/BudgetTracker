```python
#Eine leere Liste anlegen
ausgaben = []

#Neue Ausgaben hinzufügen
ausgaben.append(450) #Miete
ausgaben.append(70) #Sprit
ausgaben.append(40) #Steuern fürs Auto
ausgaben.append(150) #Einkäufe
ausgaben.append(40) #Kulturangebote

#Statistik aufstellen
gesamt = sum(ausgaben)
durchschnitt = gesamt / len(ausgaben)

#Ausgabe der Informationen
print("Ein Überblick über alle Ausgaben:".strip())
for ausgabe in ausgaben:
  print(f"- {ausgabe} €")

print("\n")
print(f"\nDeine höchste Ausgabe beträgt: {max(ausgaben)}€".strip())
print(f"Deine kleinste Ausgabe beträgt: {min(ausgaben)}€".strip())
print(f"Deine Gesamtausgaben betragen : {gesamt}€".strip())
print(f"Deine durchschnittliche Ausgabe beträgt: {durchschnitt}€".strip())
```
